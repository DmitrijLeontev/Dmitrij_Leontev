Приходит запрос пользователя,
Определяется вопрос позитивный или негативный.

Если негативный, то как Джокер.

Импорт библиотек для работы с GPT-2



```python
# Импорт библиотек
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

Инициализация модели и токенизатора

 В этой ячейке мы загружаем модель GPT-2 и её токенизатор, чтобы использовать их для генерации текстов.

```python
# Инициализация модели и токенизатора для русского языка
tokenizer = AutoTokenizer.from_pretrained('sberbank-ai/rugpt3small_based_on_gpt2')
model = AutoModelForCausalLM.from_pretrained('sberbank-ai/rugpt3small_based_on_gpt2')
```

Функция для генерации ответа

Эта функция принимает запрос и определяет, будет ли ответ от Бэтмена или Джокера. Она генерирует текст на основе этого запроса.

```python
# Функция для генерации ответа от Бэтмена или Джокера на русском языке
def generate_response(prompt, persona='batman'):
    if persona == 'batman':
        prompt = "Бэтмен: " + prompt
    elif persona == 'joker':
        prompt = "Джокер: " + prompt

    inputs = tokenizer.encode(prompt, return_tensors='pt')
    outputs = model.generate(inputs, max_length=50, num_return_sequences=1)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

Функция для определения настроения

Функция анализирует текст запроса и определяет, позитивный он или негативный, на основе ключевых слов.

```python
# Функция для определения настроения запроса на русском языке
def determine_sentiment(text):
    positive_keywords = ['хорошо', 'отлично', 'прекрасно', 'счастлив', 'радость']
    negative_keywords = ['плохо', 'грустно', 'ужасно', 'злой', 'ненавижу']
    text = text.lower()

    if any(word in text for word in positive_keywords):
        return 'batman'
    elif any(word in text for word in negative_keywords):
        return 'joker'
    else:
        return 'batman'  # По умолчанию, если не определено

```

Интерактивный чат

Эта ячейка запускает интерактивный чат, где пользователь может вводить вопросы и получать ответы от Бэтмена или Джокера. Чтобы выйти из чата, нужно ввести 'exit'.

```python
#1 Интерактивный ввод и вывод на русском языке
def interactive_chat():
    print("Добро пожаловать в интерактивный чат с Бэтменом и Джокером!")
    print("Введите 'exit' для завершения разговора.")

    while True:
        user_input = input("Вы: ")

        if user_input.lower() == 'exit':
            break

        persona = determine_sentiment(user_input)
        response = generate_response(user_input, persona=persona)

        print(f"{persona.capitalize()}: {response}")

# Запуск интерактивного чата
interactive_chat()
```
