### Задача 1

Дан следующий исходный график:

```python
import matplotlib.pyplot as plt           # Импорт модуля pyplot
import numpy as np                        # Импорт библиотеки numpy
x = np.linspace(0, 20, 101)               # Генерация последовательности из 101 числа от 0 до 20
y = np.cos(x)                             # Запись cos(x) в переменную y

plt.plot(x, y)                            # Построение графика x, y
plt.show()                                # Фиксация графика
```

Измените отображение графика таким образом, чтобы получился такой результат:

Graphic.jpg

- размер полотна: 12,4
- ширина линии: 5
- цвет линии: **4b1234**
- тип маркера: h
- размер маркера: 10
- цвет маркера: **ff00d5**
- периодичность маркера: 5
- легенда графика закреплена в правом нижнем углу: **cos(x)**
- заголовок графика: **cos(x)**
- подписи осей: **ось X** и **ось Y**
- отрисована сетка

```python
# Ваше решение
plt.figure(figsize=(12, 4))       # Указание размера полотна в дюймах

# Построение графика
plt.plot(x, y, linewidth=5, color='#4b1234', marker='h', markersize=10, markerfacecolor='#ff00d5', markevery=5, label='cos(x)')

plt.title('cos(x)')               # Заголовок графика

plt.xlabel('ось X')               # Подпись оси х
plt.ylabel('ось Y')               # Подпись оси y

plt.grid()                        # Отрисовка сетки на полотне
plt.legend(loc='lower right')     # Отображение легенды на графике в правом нижнем углу
plt.show()                        # Фиксация графика
```

### Задача 3

Создайте резервную копию датафрейма с помощью специального метода. Удалите из датафрейма все строки, где есть хотя бы один пропуск.

Затем сохраните в переменную все уникальные категории приложений (столбец **Category**). Используя цикл, вычислите количество приложений для каждой категории.

```python
# Ваше решение
df_backup = df.copy() # Создание резервной копии
```

```python
df = df.dropna(how='any') # Удаление всех строк, в которых есть хоть один пропуск
df
```

```python
unique_categories = df['Category'].unique() # Запись в переменную unique_categories уникальных категорий приложений
```

```python
for category in unique_categories:                                                    # Цикл проходится по всем элементам списка unique_categories, переменная category хранит категорию
  df_2 = df[df['Category'] == category]                                               # df_2 хранит в себе все строки той категории, на которой находится цикл
  num_rows = df_2.shape[0]                                                            # В переменной num_rows хранится количество строк получившегося датафрейма
  print('Количество приложений в категории "', category, '": ', num_rows, sep='')     # Вывод на экран категории и количества приложений (количества строк)
```
