# **Задача 1**

Создайте двумерный массив размерностью 3 на 4, состоящий из случайных целых чисел от 15 до 37.

Требуется создать новый массив той же размерности (3,4), значения которого будут зависеть от значений исходного массива и могут принимать одно из трех возможных значений:

"small", для значений исходного массива меньше 20;

"medium", для значений исходного массива в промежутке [20, 30];

"large", для значений исходного массива больше 30.
Пример:

Исходный массив:

    [[18 34 27 36]
     [16 25 24 29]
     [16 18 36 16]]
Реузльтирующий массив должен выглядеть следующим образом:

    [['small' 'large' 'medium' 'large']
     ['small' 'medium' 'medium' 'medium']
     ['small' 'small' 'large' 'small']]


```python
rand_array = np.random.randint(15, 38, (3,4))                  # Создание массива из случайных целых чисел от 15 до 37 размерностью 3 на 4
print(rand_array)                                              # Отображение созданного массива
print()                                                        # пропуск строки
new_array = np.full((3,4), 'medium')                           # Создание массива размерности  3 на 4, заполненного словом 'medium'
new_array[rand_array < 20] = 'small'                           # Изменение значений в массиве на 'small' для элементов, меньших 20
new_array[rand_array > 30] = 'large'                           # Изменение значений в массиве на 'large' для элементов, больших 30
print(new_array)                                               # Вывод сформированного массива
```

# **Задача 3**

Реализуйте алгоритм сортировки массива пузырьком.

Сортировка пузырьком - это метод сортировки массивов и списков путем последовательного сравнения и обмена соседних элементов, если предшествующий оказывается больше последующего. В процессе выполнения данного алгоритма элементы с большими значениями оказываются в конце списка, а элементы с меньшими значениями постепенно перемещаются по направлению к началу списка. Образно говоря, тяжелые элементы падают на дно, а легкие медленно всплывают подобно пузырькам воздуха.

Ваш результат может выглядеть так:

Сгенерированный массив: [71, 50, 21, 1, 81, 82, 4, 67, 76, 36]

Отсортированный массив: [1, 4, 21, 36, 50, 67, 71,76, 81, 82]

```python
Number = 10                                                         # Количество элементов в массиве
cavity = []                                                         # Создание пустого массива
for i in range(Number):                                             # Перебор элементов массива
    cavity.append(random.randint(1, 99))                            # Добавление элемента от 1 до 99 в массив
print("Сгенерированный массив: ", cavity)                           # Вывод получившегося массива
for i in range(Number-1):                                           # Перебор пар элементов
    for j in range(Number-i-1):
        if cavity[j] > cavity[j+1]:                                 # Если предшествующий больше текущего элемента
            cavity[j], cavity[j+1] = cavity[j+1], cavity[j]         # Перестановка мест пары элементов

print("Отсортированный массив: ", cavity)
```

# **Задача 4**

**Нарисуйте график рассеяния со случайным распределением по осям X и Y.**

При решении задания вам может пригодиться метод plt.scatter()


```python
X = random.sample(range(1000), 200)                 # Создание двух векторов случайных чисел
Y = random.sample(range(1000), 200)
plt.scatter(X, Y, color='r')                        # Отображение графика рассеяния
plt.xlabel("X")                                     # Подпись оси Х
plt.ylabel("Y")                                     # Подпись оси Y
plt.title("График рассеяния")                       # Заголовок графика
plt.show()
```

# **Задача 5**

Даны два списка x и y

    x = [-7, -5, 7, 9, 10, 10, 9, -7, 0, 5, 7, 4, 0, 6, 8, 4, 0]
    y = [0, 2, 2, 5, 5, 1, 0, 0, 2, 6, 6, 2, 1, -3, -3, 1, 1]
Постройте два графика: plot(x,y) и scatter(x,y) на разных координатных плоскостях.

Постройте еще один график plot(x,y) размером (6, 3) со следущими элементами:

линиями сетки;
подписями осей х и y;
названием графика "Самолет".

```python
## Данные для построения графика
x = [-7, -5, 7, 9, 10, 10, 9, -7, 0, 5, 7, 4, 0, 6, 8, 4, 0]

y = [0, 2, 2, 5, 5, 1, 0, 0, 2, 6, 6, 2, 1, -3, -3, 1, 1]
```

```python
# Импорт библиотеки для работы с графиками
import matplotlib.pyplot as plt# Построение графика plot
plt.plot(x, y)
plt.show()
```

```python
# Построение графика scatter
plt.scatter(x, y)
plt.show()
```

```python
# Назначение размера полотна графика
plt.subplots(figsize=(6, 3))

# Построение графика plot
plt.plot(x, y)

# Назначение названия графика
plt.title('Самолет')

# Подписи осей
plt.xlabel('ось х')
plt.ylabel('ось y')

# Добавление линий сетки
plt.grid()

plt.show()
```
