Данная библиотека находится в свободном доступе, использование бесплатно. Документация по ссылке: https://gtts.readthedocs.io/en/latest/

```python
# Импорт библиотеки google-cloud-texttospeech
import gtts

# Основной интерфейс для перевода текста в аудио
from gtts import gTTS

# Используем для проигрывания аудио
from IPython.display import Audio
```

Словарь со всеми доступными языками. К сожалению, для русского языка доступен только один голос:

```python
gtts.lang.tts_langs()
```

```python
# Запишем диалог оригинала в переменную

text = """-  Добрый день!
-- Привет! Я могу помочь?
-  Извините, я не могу найти расписание занятий.
-- А! Расписание занятий! Оно висит около кабинета методистов.
-  Спасибо. А не могли бы Вы мне показать, где это?
-- Конечно! Давай перейдем «на ты»! Меня зовут Анна. А тебя?
-  Хорошо. Меня зовут Кэт. Приятно познакомиться.
-- Взаимно. А вот и расписание!
-  Да, спасибо. О! У меня так много предметов! Ой, у меня же сейчас занятие! Я опаздываю! Анна, где 315 аудитория?
-- Это на третьем этаже. Беги быстрее! Удачи.
-  Тебе тоже удачи! Спасибо, что помогла.
-- Не за что. До встречи.
-  Еще раз большое спасибо. До встречи!"""

# Преобразовываем текст в аудио
tts = gTTS(text, lang='ru')

# Сохраняем в формате мр3
tts.save('blok.mp3')
```

```python
# Прослушаем полученное аудио оригинал
Audio('blok.mp3')
```
