```python
!pip install openai
!pip install gtts
import openai
import os
from gtts import gTTS
from IPython.display import Audio
from getpass import getpass
from openai import OpenAI
```

```python
# # запустите эту ячейку, если используете секретный ключ в колабе
from google.colab import userdata
key = userdata.get('OPENAI_API_KEY')
os.environ["OPENAI_API_KEY"] = key
```

```python
# Читаем в оригинале и задаем system промпт: "Ты - старательный помощник. Прочитай вслух стих. Ничего от себя не добавляй."
def chat_with_gpt_to_audio(user_input):
    # Инициализация клиента OpenAI
    client = OpenAI()
    # Создание запроса к модели
    result = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Ты - старательный помощник. Прочитай вслух стих. Ничего от себя не добавляй."},
            {"role": "user", "content": user_input}
        ]
    )

    # Получение ответа модели
    message = result.choices[0].message.content

    # Преобразование текста в аудио
    tts = gTTS(text=message, lang='ru')
    #audio_file = "output.mp3"
    #tts.save(audio_file)
    tts.save("output.mp3")
    display(Audio("output.mp3"))
    return

# Пример использования функции
user_input = "Наша Таня громко плачет: Уронила в речку мячик. Тише Танечка, не плачь:Не утонет в речке мяч."
audio_path = chat_with_gpt_to_audio(user_input)
```

```python
# Импорт модуля
import elevenlabs
from elevenlabs import play, stream, save
from elevenlabs.client import ElevenLabs
```

```python
import getpass  # Импорт модуля для скрытия ввода пароля

# Получение ключа API от пользователя с помощью функции input
api_key = getpass.getpass(prompt="Введите ключ API для elevenlabs: ")

# Установка ключа API для библиотеки elevenlabs
client = ElevenLabs(api_key=api_key)
```

Обратимся к модели elevenlabs и сгенерируем речь из простенького текста

```python
# Генерация аудиофайла с использованием указанного текста, голоса и модели
audio = client.generate(
    text="Наша Таня громко плачет: Уронила в речку мячик. Тише Танечка, не плачь:Не утонет в речке мяч. ",  # Текст, который нужно преобразовать в речь
    voice="Charlie",                # имя голоса, который будет использован
    model="eleven_multilingual_v2"  # Модель, используемая для генерации речи
)

# Воспроизведение сгенерированного аудио
play(audio, notebook=True)
```

Довольно хорошо генерирует. Как живой человек.

```python
# Сохранение сгенерированного аудио в файл "audio.mp3"
save(audio, "audio.mp3")
```

Синтезируем айдиофайл из ответа GPT:

```python
# Импортируем модуль для расчета времени выполнения функции
from datetime import datetime
```

```python
def chat_with_gpt_to_audio(user_input):
    start_time = datetime.now()  # Засекаем время начала выполнения функции
    # Инициализация клиента OpenAI
    openai_client = OpenAI()
    # Создание запроса к модели
    result = openai_client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Ты - старательный помощник. Прочитай этот стих в разговорном стиле дословно и затем обьясни его содержание."},
            {"role": "user", "content": user_input}
        ]
    )

    # Получение ответа модели
    message = result.choices[0].message.content

    # Генерация аудиофайла с использованием elevenlabs
    voice = elevenlabs.Voice(
        voice_id="21m00Tcm4TlvDq8ikWAM",
        settings=elevenlabs.VoiceSettings(
            stability=0,
            similarity_boost=0.75
        )
    )
    audio = client.generate(
        text=message,
        voice=voice,
        model="eleven_multilingual_v2"
    )

    # Воспроизведение сгенерированного аудио на ноутбуке
    play(audio, notebook=True)

    # Сохранение сгенерированного аудио в файл "output.mp3"
    save(audio, "output.mp3")

    end_time = datetime.now()  # Засекаем время завершения выполнения функции
    execution_time = end_time - start_time  # Рассчитываем время выполнения
    print("Время выполнения:", execution_time)  # Выводим время выполнения
    return

# Пример использования функции
user_input = "Наша Таня громко плачет: Уронила в речку мячик. Тише Танечка, не плачь:Не утонет в речке мяч."
chat_with_gpt_to_audio(user_input)
```
