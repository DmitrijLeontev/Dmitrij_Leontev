# Обращение к chatGPT через API.


OpenAI обеспечивает доступ к своему языковому моделированию AI, включая ChatGPT, через простой и мощный API. Это позволяет разработчикам встраивать эту технологию в собственные приложения или услуги, создавая решения, которые используют передовые возможности генерации текста.

**Что такое API?**
API, или интерфейс программирования приложений, - это набор процедур и функций, которые позволяют одному программному обеспечению взаимодействовать с другим. API для ChatGPT от OpenAI позволяет разработчикам отправлять запросы на генерацию текста и получать ответы напрямую из своих программных приложений.

Если в очень простом формате, то API - это как почтальон в мире компьютеров. Когда вы хотите что-то спросить у ChatGPT (это большой умный робот, который может говорить и отвечать на вопросы), вы пишете свой вопрос на листе бумаги и даете его этому почтальону (API).

Почтальон бежит к ChatGPT, дает ему лист бумаги с вашим вопросом и ждет, пока ChatGPT напишет ответ. Когда ответ готов, почтальон берет его и бежит обратно к вам.

Поэтому, когда мы говорим "обращение к ChatGPT через API", мы имеем в виду, что мы отправляем наш вопрос почтальону, который доставляет его ChatGPT и принимает обратно ответ для нас. Это очень удобно, потому что мы можем задавать вопросы откуда угодно и получать ответы быстро.

# Роли chatGPT.  


Давайте поговорим о том, как общаться с моделью от OpenAI, такой как "gpt-3.5-turbo", используя их API.

ChatGPT использует идею "ролей". Каждое сообщение в списке сообщений - это набор данных, включающий два элемента: role и content. Role может быть одной из трех вещей: "system", "user" или "assistant", указывая, кто в данный момент "разговаривает". Content - это сам текст сообщения.

Обычно, в диалоге участвуют три главные роли:

**"system":** Роль системы используется для задания общей обстановки разговора. Это своего рода руководство для модели, указывающее, как она должна вести себя в процессе диалога.

**"user":** Это роль человека, который ведет диалог с моделью. Ваши вопросы модели будут представлены под этой ролью. Также сюда мы можем подавать определенные инструкции для модели (самые основные и важные),  при необходимости - содержание предыдущего диалога, найденные релевантные отрезки текста (об этом подробно будем говорить на следующем занятии).

**"assistant":** Это роль модели AI, которая отвечает на вопросы пользователя. То есть, при необходимости, мы можем "подсказать" модели, какой ответ мы от нее ожидаем получить, подав его в assistant.

```python
import getpass
import openai
import os
# Получение ключа API от пользователя и установка его как переменной окружения
openai_key = getpass.getpass("OpenAI API Key:")
os.environ["OPENAI_API_KEY"] = openai_key
openai.api_key = openai_key
```

```python
# Теперь вы можете взаимодействовать с API OpenAI
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "Ты- старательный помощник."},
        {"role": "user", "content": "Какое отношение имеет Пушкин к искусственному интеллекту?"}
    ]
)

print(result)  # вывод результата запроса
message = result['choices'][0]['message']['content']
print(message)
```

Ответ (result) возвращается в формате JSON.

Чтобы получить ответ на русском языке, мы можем извлечь содержимое ответа из JSON и распечатать его (как это сделано в примере выше). Давайте попробуем поменять содержание "system" и посмотрим, как это изменит суть ответа:

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "Ты- человек из прошлого века, отвечаешь в соответствующем стиле и с использованием старо-русской лексики"},
        {"role": "user", "content": "Какое отношение имеет Пушкин к искусственному интеллекту?"}
    ]
)

print(result)  # вывод результата запроса
message = result['choices'][0]['message']['content']
print(message)
```

# Что такое промт и его влияние на результат

"Промпт" в контексте использования API для моделей на основе GPT относится к строке текста, которую вы передаете модели в качестве входных данных. Вы можете думать о промпте как о "запросе" или "вводе", который модель использует, чтобы сгенерировать свой выходной текст.

Важность промпта состоит в том, что он определяет, как модель сгенерирует свой ответ. Промпт служит начальной точкой для модели и указывает ей, в каком направлении вы хотите, чтобы генерация текста продолжалась.

То есть, как мы увидели в примерах выше, от промпта зависит, как модель будет отвечать на переданный ей вопрос: указав "Ты- старательный помощник." - мы получаем наиболее нейтральный и содержательный ответ по существу, однако, изменив промпт на "Ты- человек из прошлого века, отвечаешь в соответствующем стиле и с использованием старо-русской лексики" - мы получаем ответ с использованием соответствующего стиля и древне-русской лексики.

Давайте проведем еще пару экспериментов, чтобы лучше понять, как промпт влияет на ответ модели:

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "Ты- поэт, отвечаешь только стихами"},
        {"role": "user", "content": "Какое отношение имеет Пушкин к искусственному интеллекту?"}
    ]
)

print(result)  # вывод результата запроса
message = result['choices'][0]['message']['content']
print(message)
```

Какие еще параметры можно использовать, чтобы управлять ответами модели:

* **max_tokens**: Максимальное количество токенов в ответе. Это помогает контролировать длину генерируемого текста.

* **temperature**: Параметр, который контролирует степень случайности генерируемого текста. Более высокие значения приводят к более случайному тексту, в то время как более низкие значения делают генерацию более детерминированной.

* **n:** Количество независимых ответов, которые должна сгенерировать модель.


```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "Ты- научный сотрудник, оперируешь только фактами, отвечаешь с использованием научных терминов, кратко и только по сути"},
        {"role": "user", "content": "Какое отношение имеет Пушкин к искусственному интеллекту?"}
    ],
  max_tokens=100,
  temperature=0.5,
  n=1
  )

print(result)  # вывод результата запроса
message = result['choices'][0]['message']['content']
print(message)
```

Как можно заметить, генерация ответа моделью прервалась из-за достижения в ответе количества токенов, указанных в параметре max_tokens. Попробуем изменить параметры вызова функции и оценим разницу в результате:

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "Ты- научный сотрудник, оперируешь только фактами, отвечаешь с использованием научных терминов, кратко и только по сути"},
        {"role": "user", "content": "Какое отношение имеет Пушкин к искусственному интеллекту?"}
    ],
  max_tokens=300,
  temperature=1,
  n=3
)

print(result)  # вывод результата запроса
message_0 = result['choices'][0]['message']['content']
print("Первый ответ модели:")
print(message_0)
print("Второй ответ модели:")
message_1 = result['choices'][1]['message']['content']
print(message_1)
print("Третий ответ модели:")
message_2 = result['choices'][2]['message']['content']
print(message_2)
```

# Токены и их подсчет.

Токены являются фундаментальными элементами, с которыми работают модели GPT. Вместо обработки полных слов, модель декомпозирует текст на более мелкие сегменты, известные как токены. Это могут быть отдельные буквы, сложные слова или целые слова, в зависимости от языка и контекста. Для английского языка большая часть слов состоит из одного или двух токенов. Например, слово "ChatGPT" разбивается на два токена: "Chat" и "GPT". В русском языке ситуация отличается - 1 токен не всегда соответствует 1 слову, скорее, 1 токен соответствует 1 слогу.

Ограничение на количество токенов связано с максимальным числом токенов, которое модель способна обработать за одну итерацию. Для gpt-3.5-turbo это число составляет 4096 токенов. Это значит, что общее количество токенов в ваших входных данных и ответах модели не должно превышать 4096.

Недавно была представлена модель gpt-3.5-turbo-16k. Эта модель предоставляет в четыре раза больше контекста, чем базовая модель с ограничением в 4k токенов.

Поскольку количество токенов влияет на стоимость и время выполнения запроса, важно контролировать количество токенов в ваших запросах. Если ваш запрос превышает максимальное ограничение по токенам, он может быть отклонен. Кроме того, если входные данные содержат слишком много токенов, то может не хватить токенов для получения полезного ответа от модели. Поэтому иногда необходимо управлять количеством токенов в ваших входных данных.

Когда дело доходит до коротких запросов, как те, которые мы делали ранее, проблем не возникает, мы не превышаем лимит по токенам. Однако, когда нам нужно подавать на вход модели большие блоки текста, подсчет токенов становится критически важным. Давайте разберемся, как можно легко подсчитать количество токенов в вводимых текстах. Для начала, создадим переменные, в которых будут находится тексты, которые мы будем подавать в system и user:

```python
system = '''Ты менеджер по продажам в чате компании, компания продает курсы по AI. У тебя есть документ со всеми материалами о продуктах компании.
Тебе задает вопрос клиент в чате, дай ему ответ, опираясь на документ, постарайся ответить так, чтобы человек захотел после ответа купить обучение.
Отвечай максимально точно по документу, не придумывай ничего от себя. Не упоминай документ при ответе.'''

base_info= '''Курс по AutoML состоит из пяти занятий, каждое из которых посвящено конкретным темам:Создание моделей на AutoKeras.
На этом занятии будет полный разбор фреймворка AutoKeras, включая все Подробно разберем всё на AutoKeras, все гриды, тюнеры и другие особенности.
Оптимизация моделей на KerasTuner и Talos. Методы подбора параметров сетки для заданной архитектуры модели на фреймворках.
Библиотеки для построения генетических алгоритмов. Аналог AutoKeras. Как пользоваться и как в целом устроен.
Оптимизация сложных моделей на KerasTuner и Talos. Углубление в тему оптимизации архитектур, которые AutoKeras  не может создавать, на фреймворках KerasTuner и Talos.
Будут рассмотрены примеры, включая построение U-Net для сегментации изображений.
Свободная тема на выбор группы. Студенты сами выберут тему из предложенных на курсе. Например разбор генетических алгоритмов, Auto-sklearn или Light AutoML, или их комбинацию.
Как получить курс:
Курс не доступен для покупки, а предоставляется в качестве подарка новичкам, приобретающим любой тариф, кроме базового.
Также курс доступен и для старичков, которые могут получить его в подарок при любой покупке на сумму 40 000 рублей.
К примеру, это может быть покупка курса по PyTorch или Интеграции в Production, стажировок на год, либо повышение тарифа с базового на основной или продвинутый.'''

question = "Сколько занятий в курсе по AutoML"
```
